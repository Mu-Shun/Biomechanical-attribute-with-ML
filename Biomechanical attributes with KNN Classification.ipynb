Vertebral Column Data Set
This Biomedical data set was built by Dr. Henrique da Mota during a medical residence
period in Lyon, France. Each patient in the data set is represented in the data set
by six biomechanical attributes derived from the shape and orientation of the pelvis
and lumbar spine (in this order): pelvic incidence, pelvic tilt, lumbar lordosis angle,
sacral slope, pelvic radius and grade of spondylolisthesis. The following convention is
used for the class labels: DH (Disk Hernia), Spondylolisthesis (SL), Normal (NO) and
Abnormal (AB). In this exercise, we only focus on a binary classification task NO=0
and AB=1

### iii Since the computation time depends on the size of the training set, one may only use a subset of the training set. ### i-B. with log10(p) âˆˆ {0.1, 0.2 ,0.3 , . . . ,1}. In this case, use the k âˆ— you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?

```python
log_p = []
error_log_p = 100
k_star_log_p = 0
for i in range(1,11):
    log_p.append(10 ** (i/10))
for j in log_p:
    neigh_Manhattan_Distance = KNeighborsClassifier(n_neighbors = 6, p = j)
    neigh_Manhattan_Distance.fit(training_set, training_label) 
    neigh_Manhattan_Distance.predict(test_set)
    if sum(np.power(neigh_Manhattan_Distance.predict(test_set)-(test_label),2)) < error_log_p:
        error_log_p = sum(np.power(neigh_Manhattan_Distance.predict(test_set)-(test_label),2))
        k_star_log_p = j  
print("Best log(10)p :", k_star_log_p, "which is 10**0.6")
print("Test Error numbers :", error_log_p)
```

### i-C. which becomes Chebyshev Distance with p â†’ infinite

```python
k_star_Chebyshev = 0
error_Chebyshev = 100
for i in k_Manhattan_Distance:
    neigh_Manhattan_Distance = KNeighborsClassifier(n_neighbors = i, p = float("inf"))
    neigh_Manhattan_Distance.fit(training_set, training_label) 
    neigh_Manhattan_Distance.predict(test_set)
    if sum(np.power(neigh_Manhattan_Distance.predict(test_set)-(test_label),2)) < error_Chebyshev:
        error_Chebyshev = sum(np.power(neigh_Manhattan_Distance.predict(test_set)-(test_label),2))
        k_star_Chebyshev = i
        
print("Best k :", k_star_Chebyshev)
print("Test Error numbers :", error_Chebyshev)
```

## (e) The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is inversely proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away. Use weighted voting with Euclidean, Manhattan, and Chebyshev distances and report the best test errors when k âˆˆ {1, 6, 11, 16, . . . ,196}

```python
k_Weight_Distance = []
n_Weight_Distance = 1
while n_Weight_Distance < 200:
    k_Weight_Distance.append(n_Weight_Distance)
    n_Weight_Distance = n_Weight_Distance + 5
test_error_Weight_Distance = []
p_Weight_Distance = [1, 2, float('inf')]
suitable_k_Weight_Distance = 0
for i in p_Weight_Distance:
    error_numbers_Weight_Distance = 100
    for j in k_Weight_Distance:
        neigh_Weight_Distance = KNeighborsClassifier(n_neighbors = j, weights='distance', p = i)
        neigh_Weight_Distance.fit(training_set, training_label) 
        test_error_Weight_Distance.append(sum(np.power(neigh_Weight_Distance.predict(test_set)-(test_label),2)))
        if sum(np.power(neigh_Weight_Distance.predict(test_set)-(test_label),2)) < error_numbers_Weight_Distance:
            error_numbers_Weight_Distance = sum(np.power(neigh_Weight_Distance.predict(test_set)-(test_label),2))
            suitable_k_Weight_Distance = j
            k_star_predict_Weight_Distance = neigh_Weight_Distance.predict(test_set)
    print("When p = ",i,",the suitable k is :", suitable_k_Weight_Distance, ",and the error number is", error_numbers_Weight_Distance)
```
